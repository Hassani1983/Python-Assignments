### Create two operands and play around using different operators

```python
x = 10
y = 3

# Python Arithmetic Operators
print("Python Arithmetic Operators")
r = x + y
print('x + y ==> ',r)
r = x - y
print('x - y ==> ',r)
r = x * y
print('x * y ==> ',r)
r = x / y
print('x / y ==> ',r)
r = x % y
print('x % y ==> ',r)
r = x ** y
print('x ** y ==> ',r)
r = x // y
print('x // y ==> ',r)
print("")

# Python Comparison Operators
print("Python Comparison Operators")
print('x == y ? ==> ',x == y)
print('x != y ? ==> ',x != y)
print('x > y ? ==> ',x > y)
print('x < y ? ==> ',x < y)
print('x >= y ? ==> ',x >= y)
print('x <= y ? ==> ',x <= y)
print("")

# Python Assignment Operators
print("Python Assignment Operators")
x = 5
print("x = 5")
x += 3
print("x += 3 ==> ",x)
x -= 3
print("x -= 3 ==> ", x)
x *= 3
print("x *= 3 ==> ", x)
x /= 3
print("x /= 3 ==> ", x)
x %= 3
print("x %= 3 ==> ", x)
x //= 3
print("x //= 3 ==> ", x)
x = 5
x **= 3
print("x **= 3 ==> ", x)
x = 5
x &= 3
print("Bitwise And x &= 3 ==> ", x)
x = 5
x |= 3
print("Bitwise OR x |= 3 ==> ", x)
x = 5
print("Bitwise Not ~x ==> ", ~x)
x = 5
x ^= 3
print("Bitwise XOR x ^= 3 ==> ", x)
x = 5
x >>=3
print("Bitwise Shift Right x >>= 3 ==> ", x)
x = 5
x<<=3
print("Bitwise Shift Left x <<= 3 ==> ", x)
print()

# Python Logical Operators
print("Python Logical Operators")
print("x < 5 and  x < 10 ==> ",x < 5 and  x < 10)
print("x < 5 or x < 4 ==> ",x < 5 or x < 4)
print("not(x < 5 and x < 10) ==> ",not(x < 5 and x < 10))
print()

x = "banana"
y = ["apple", "banana"]
# Python Identity Operators
print("Python Identity Operators")
print("x is y ==> ",x is y)
print("x is not y ==> ",x is not y)
print()

# Python Membership Operators
print("Python Membership Operators")
print("x in y ==> ",x in y)
print("x not in y ==> ",x not in y)
```

```python
a = int(10)
constant_val = 3
for dm in dir(a):
    command = str('a.') + str(dm) + str('(') + str(constant_val) + str(')')
    try:
        print(command," ==> ", eval(command))
    except:
        command = str('a.') + str(dm) + str('()')
        try:
            print(command," ==> ", eval(command))
        except:
            print("Error in {}".format(command))
```

#### Learn different ways to format the string. 
#### Tutorials on https://realpython.com/python-formatted-output/
#### Try: 

```python

print("I love", "python") 
print("I love" + "python") 
print("I love"+" python" + "version", 3.10)
print("I love"+" python" + "version %.2f"%3.10)
print("I love %s version %.2f" % ("python", 3.10))
# Using format positional argument
print("I love {} version {}".format("python", 3.10))
print("I love {} version {}".format("python", "3.10")) 
print("I love {} version {:.2f}".format("python", 3.10)) 
print("I love {1} version of {0}".format("python", "3.10"))
# Using format keyword argument
print("I love {pl} version {version: .2f}".format(pl="python", version=3.10))
# Using f-string literal
info={"pl": "python", "version":3.10}
print(f"I love {info.get('pl')} version {info.get('version'): .2f}")
# Can be done with uppercase f literal and single quote but double quote while accessing dictionary value
print (F'I love {info["pl"]} version {info["version"]: .2f}')
#f-string conditional
batch='ML'
print(f'I love {"javascript" if batch=="FS" else "python"}')
```

```python
countries = ["Finland","Sweden","Denmark","Netherlands","Germany","Norway","Iceland"]
print("countries ==> ",countries)
print("Find the index of “Sweden” ==> ",countries.index("Sweden"))
print("Get all the items except the first ==> ",countries[1:])
print("Get all the items except the last ==> ",countries[:-1])
print("Find the last item using negative indexing ==> ",countries[-1])
print("Find the last three item using negative indexing ==> ",countries[-3:])
print("Grab all the items with odd indices ==> ",countries[1::2])
print("Find the first and last item for the list ==> ",countries[0],countries[-1])
print("Get [”Netherlands”,”Germany”,”Norway”] using negative indexing ==> ",countries[-4:-1])
print("What is the output of countries[::-2] ==> ",countries[::-2])
print("What is the output of countries[1:100:2]? ==> ",countries[1:100:2])
print("What is the output of countries[len(countries)-1:0:2]? ==> ",countries[len(countries)-1:0:2])
print("Get the output ['Germany', 'Netherlands', 'Denmark'] ==> ",countries[-3:-6:-1])
print("Use enumerate method to get index and values in the list ==> ")
for i,v in enumerate(countries):
    print(i," : ", v)
countries.reverse()
print("Reverse the list ==> ",countries)
```
