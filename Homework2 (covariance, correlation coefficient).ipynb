```python
s1 = df['S1'].sample(50)
s2 = df['S2'].sample(50)
s3 = np.random.choice(df['S3'] ,50 , replace=False)
```

```python
# Confidence level: [90%, 95%, 99%]
Z_value = [1.64, 1.96, 2.58]
mean = [np.mean(s1), np.mean(s2), np.mean(s3)]
std = [np.std(s1), np.std(s2), np.std(s3)]
print('Z_value ==> ',Z_value)
print('mean ==> ',mean)
print('std ==> ',std)
```

```python
LCL = []
UCL = []

for i in range(3):
    LCL_s = []
    UCL_s = []
    for z in range(3):
        LCL_s.append(mean[i]-Z_value[z]*(std[i]/np.sqrt(50)))
        UCL_s.append(mean[i]+Z_value[z]*(std[i]/np.sqrt(50)))
    LCL.append([round(x,2) for x in LCL_s])
    UCL.append([round(x,2) for x in UCL_s])    
    
print("Lower Confidence Value ==> ",LCL)
print("Uper  Confidence Value ==> ",UCL)
```

```python
sample_mean = [np.mean(s1), np.mean(s2), np.mean(s3)]
population_mean = [np.mean(df['S1']), np.mean(df['S2']), np.mean(df['S3'])]
print("sample_mean     : ",sample_mean)
print("population_mean : ", [round(x,2) for x in population_mean])
```
